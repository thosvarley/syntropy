```python
import pid
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

We begin by loading in the data, it's 10 channels from an fMRI BOLD scan.

```python
data = pd.read_csv("bold.csv", header=None).values
```

A quick visualization to get an intiuitive sense of what we're dealing with. 

```python
fig = plt.figure(figsize=(8.5,4), dpi=100)
ax = fig.add_subplot(1,1,1)

for i in range(data.shape[0]):
    _ = ax.plot(data[i]+(5*i), color="k")
    
_ = ax.set_yticks([i*5 for i in range(data.shape[0])],
                  [i+1 for i in range(data.shape[0])])
ax.set_ylabel("Channel index")
ax.set_xlabel("TR")
ax.set_title("BOLD data")
```

The easiest thing to do is the PID for a single source and a single target:

```python
inputs = (0,1)
target = (2,)

ptw, avg = pid.partial_information_decomposition(inputs, target, data)
```

A good sanity check is whether the sum of the partial information atoms matches the mutual information computed from the covariance matrices. 

```python
mi = pid.mutual_information(inputs, target, data)

print("Covariance MI: {0} nat".format(mi))
print("Sum PI-atoms:  {0} nat".format(sum(avg.values())))
```

We can also set a target that is the joint-state of multiple channels simultaniously.

```python
inputs = (0,1)
target = (2,3)

ptw, avg = pid.partial_information_decomposition(inputs, target, data)
mi = pid.mutual_information(inputs, target, data)

print("Covariance MI: {0} nat".format(mi))
print("Sum PI-atoms:  {0} nat".format(sum(avg.values())))
```

If you want to account for the temporal nature of the data, you can slice and dice the original time series a little bit. 

For example, lets say you want the PID of the information that two channels disclose about their own joint future:

```python
data_lag = np.vstack((data[inputs, :-1], data[inputs, 1:]))

ptw, avg = pid.partial_information_decomposition((0,1), (2,3), data_lag)

# Note that this is now TDMI - time-delayed mutual information. 
tdmi = pid.mutual_information((0,1), (2,3), data_lag)

print("Covariance MI: {0} nat".format(tdmi))
print("Sum PI-atoms:  {0} nat".format(sum(avg.values())))
```

By examining the PID, we can get interesting temporal dependencies

```python
print("Redundancy {{0}}{{1}}: {0} nat".format(avg[((0,),(1,))]))
print("Unique 0 {{0}}: {0} nat".format(avg[((0,),)]))
print("Unique 1 {{1}}: {0} nat".format(avg[((1,),)]))
print("Synergy {{0,1}}: {0} nat".format(avg[((0,1,),)]))
```

So, these random fMRI BOLD time series have almost equal amounts of redundant and synergistic predictive power about their own future, and little unique information.

We can also do a multivariate entropy decomposition instead of a multivariate information decomposition, using the h_min measure.


```python
ptw, avg = pid.partial_entropy_decomposition((0,1), data=data)

entropy = pid.joint_entropy((0,1), data=data)

print("Joint entropy: {0} nat".format(entropy))
print("Sum PE-atoms:  {0} nat".format(sum(avg.values())))
```

In the case of bivariate inputs, we can re-extract the mutual information as the difference between the redundant and synergistic atoms.

```python
ped_mi = avg[((0,),(1,))] - avg[((0,1),)]
mi = pid.mutual_information((0,),(1,),data)

print("MI from PED:  {0} nat".format(ped_mi))
print("MI from covs: {0} nat".format(mi))
```

Finally we have the generalized information decomposition, which will decompose the information gained when updating from a prior Q to a posterior P in the form of the Kullback-Leibler divergence. 

In the case where the prior is maximum entropy, this is equivalenet to the total correlation, which is a nice sanity check. 

```python
tc = pid.total_correlation(data[(0,1,2),:])

cov_prior = np.eye(data.shape[0])
cov_posterior = np.cov(data, ddof=0.0)

ptw, avg = pid.generalized_information_decomposition((0,1,2), data, cov_prior, cov_posterior)

print("TC from covs:  {0} nat".format(tc))
print("Sum PI values: {0} nat".format(sum(avg.values())))
```
